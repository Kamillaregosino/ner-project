## Convertendo de Label Studio par Spacy

```python
import json

with open("labeled2.json", "r", encoding="utf-8") as f:
    data = json.load(f)

dataset = []

for item in data:
    texto = item["data"]["text"]
    entities = []
    
    for ann in item.get("annotations", []):
        for res in ann.get("result", []):
            start = res["value"]["start"]
            end = res["value"]["end"]
            label = res["value"]["labels"][0]  
            entities.append((start, end, label))
    
    if entities:
        dataset.append((texto, {"entities": entities}))

with open("data_spacy2.py", "w", encoding="utf-8") as f:
    f.write(f"Dataset = {dataset}")
```
